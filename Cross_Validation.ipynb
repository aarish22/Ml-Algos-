```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from IPython.display import Image
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
kf = KFold(n_splits=5)
```

```python
model = LogisticRegression(max_iter=200)
```

# K-Fold Cross Validation

```python
Image('/content/k-fold-Cross-ValidationCV-Illustration.png')
```

```python
scores = cross_val_score(model, X, y, cv=kf)
print("K-Fold Cross-Validation Scores:", scores)
```

# Stratified K-Fold Cross Validation

```python
Image('/content/Visualization-of-stratified-k-fold-cross-validation-with-k5.png')
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=5)
stratified_scores = cross_val_score(model, X, y, cv=skf)
```

```python
print("Stratified K-Fold Cross-Validation Scores:", stratified_scores)
```

# Leave-One-Out Cross Validation

```python
Image('/content/1-3-1024x622.png')
```

```python
from sklearn.model_selection import LeaveOneOut
```

```python
loo = LeaveOneOut()
```

```python
loo_scores = cross_val_score(model, X, y, cv=loo)
```

```python
print("Leave-One-Out Cross-Validation Scores:", loo_scores)
```

# Leave-P-Out Cross-Validation

```python
from sklearn.model_selection import LeavePOut
```

```python
lpo = LeavePOut(p=2)
```

```python
lpo_scores = cross_val_score(model, X, y, cv=lpo)
```

```python
print("Leave-P-Out Cross-Validation Scores:", lpo_scores)
```

# Time Series Split

```python
Image('/content/An-illustration-of-our-chosen-time-series-cross-validation-strategy-The-data-is-first.png')
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
time_series_scores = cross_val_score(model, X, y, cv=tscv)
```

```python
print("Time Series Split Cross-Validation Scores:", time_series_scores)
```
