```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import squarify
```

#Line Plot
Description: A line plot displays information as a series of data points called 'markers' connected by straight line segments.

Use Case: Best for showing trends over time.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

# Bar Chart
Description: A bar chart presents categorical data with rectangular bars.

Use Case: Best for comparing quantities across categories.

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 2, 4]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

#Histogram
Description: A histogram represents the distribution of numerical data.

Use Case: Best for showing the frequency distribution of a dataset.

```python
data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

#Scatter Plot
Description: A scatter plot uses dots to represent values for two different numeric variables.

Use Case: Best for showing the relationship between two variables.

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

# Box Plot
Description: A box plot displays the distribution of data based on a five-number summary: minimum, first quartile, median, third quartile, and maximum.

Use Case: Best for showing the distribution and outliers in a dataset.

```python
data = [np.random.rand(100) for _ in range(4)]

plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=['A', 'B', 'C', 'D'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

#Pie Chart
Description: A pie chart is a circular statistical graphic divided into slices to illustrate numerical proportions.

Use Case: Best for showing the relative proportions of different categories.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
plt.title('Pie Chart')
plt.show()
```

#Heatmap
Description: A heatmap is a data visualization technique that shows the magnitude of a phenomenon(correlation) as color in two dimensions.

Use Case: Best for showing data density and variations across multiple variables.

```python
data = np.random.rand(10, 12)

plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

#Violin Plot
Description: A violin plot is similar to a box plot, but with a rotated kernel density plot on each side.

Use Case: Best for showing the distribution of data across different categories.

```python
data = [np.random.rand(100) for _ in range(4)]
labels = ['A', 'B', 'C', 'D']

plt.figure(figsize=(10, 6))
sns.violinplot(data=data)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Violin Plot')
plt.show()
```

#Pair Plot
Description: A pair plot is used to plot pairwise relationships in a dataset.

Use Case: Best for exploring relationships between multiple variables.

```python
data = pd.DataFrame(np.random.randn(100, 4), columns=['A', 'B', 'C', 'D'])

sns.pairplot(data)
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

#Area Plot
Description: An area plot displays quantitative data graphically.

Use Case: Best for showing cumulative totals over time.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))
plt.fill_between(x, y1, alpha=0.5, label='sin(x)')
plt.fill_between(x, y2, alpha=0.5, label='cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Plot')
plt.legend()
plt.show()
```

#Bubble Chart
Description: A bubble chart is a type of scatter plot where the size of the marker is proportional to a third variable.

Use Case: Best for showing the relationship between three variables.

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 1000

plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, alpha=0.5, color='purple')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Chart')
plt.show()
```

#Treemap
Description: A treemap displays hierarchical data as a set of nested rectangles.

Use Case: Best for showing the part-to-whole relationships in hierarchical data.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [500, 300, 150, 50]

plt.figure(figsize=(10, 6))
squarify.plot(sizes=sizes, label=labels, alpha=0.7)
plt.title('Treemap')
plt.axis('off')
plt.show()
```
