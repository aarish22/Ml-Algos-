```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
```

```python
wine = load_wine()
X_wine = wine.data
y_wine = wine.target
```

```python
X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(X_wine, y_wine, test_size=0.3, random_state=42, stratify=y_wine)
```

```python
bagging_wine = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42)
bagging_wine.fit(X_train_wine, y_train_wine)
y_pred_bagging_wine = bagging_wine.predict(X_test_wine)
bagging_accuracy_wine = accuracy_score(y_test_wine, y_pred_bagging_wine)
```

```python
bagging_accuracy_wine
```

```python
random_forest_wine = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_wine.fit(X_train_wine, y_train_wine)
y_pred_rf_wine = random_forest_wine.predict(X_test_wine)
rf_accuracy_wine = accuracy_score(y_test_wine, y_pred_rf_wine)
```

```python
rf_accuracy_wine
```

```python
labels = ['Bagging', 'Random Forest']
accuracies = [bagging_accuracy_wine, rf_accuracy_wine]

plt.figure(figsize=(16, 5))
plt.bar(labels, accuracies, color=['blue', 'green'])
plt.ylabel('Accuracy')
plt.title('Comparison of Bagging and Random Forest on Wine Dataset')
plt.ylim(0.7, 1.0)
for i, v in enumerate(accuracies):
    plt.text(i, v + 0.005, f"{v:.3f}", ha='center', fontweight='bold')
plt.show()
```

```python
cm_bagging_wine = confusion_matrix(y_test_wine, y_pred_bagging_wine)
cm_rf_wine = confusion_matrix(y_test_wine, y_pred_rf_wine)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.heatmap(cm_bagging_wine, annot=True, fmt='d', ax=axes[0], cmap='Blues')
axes[0].set_title('Bagging Classifier - Wine')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')


sns.heatmap(cm_rf_wine, annot=True, fmt='d', ax=axes[1], cmap='Greens')
axes[1].set_title('Random Forest Classifier - Wine')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')
```
